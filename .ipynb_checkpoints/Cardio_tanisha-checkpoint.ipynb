```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Set pandas options to display all columns (visual aid).
pd.set_option('display.max_columns',None)
```

```python
df = pd.read_csv('cardio_train.csv',sep = ";")
df
```

```python
df.describe() 
#give summary of a dataframe
```

```python
df.info()
#give infomation of dataframe
```

```python
df.isnull().sum()
#give null values
```

```python
df.isnull().sum()/df.shape[0]*100
#null value find karse and then eno sum and then row percentage divide
```

```python
df['Age_Year']=(df["age"]/365).astype(int)
df
#convert Age is in days
```

```python
df=df.drop("age",axis=1)
df
#remove age 
```

```python
df["gender"]=df["gender"].map({1:0,2:1})
df
#old values ne new values ma put kare
```

```python
cols = ['height', 'weight', 'ap_hi', 'ap_lo']

for col in cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    df = df[(df[col] >= lower) & (df[col] <= upper)]
df
```

```python
#find BMI
df['BMI']=df['weight']/((df['height']/100)**2)
df
```

```python
#To detect duplicates
df.duplicated().sum()
```

```python
## remove wrong medical data ---- outliers

## ap_lo> ap_hi --- not valid
df= df[df['ap_lo']<=df['ap_hi']]
## height <100  or >220--- unrealistic 
df=df[(df['height']>100) & (df['height']<220)]
## weight < 30 or >200
df=df[(df['weight']>30) & (df['weight']<200)]
df
```

```python
#give graphical represtation of null values
#EDA :- Exploratory Data Analysis
sns.heatmap(df.corr(), cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

<h1> Data Training </h1>

```python
from  sklearn.model_selection import train_test_split
## seperate input and output
## x = testing data
## y = target data

x = df.drop(columns=['cardio'],axis=1)  ## all data except cardio column is training data
y = df['cardio']  ## cardio is ourtarget data to check the correctness of model

x_train, x_test , y_train, y_test = train_test_split(
    x,y,test_size=0.2 , random_state=42,
)
```

```python
x_train
```

```python
df
```

```python
num_cols=['height','weight','ap_hi','ap_lo','BMI']
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train[num_cols])
```

```python
x_train[num_cols]=scaler.transform(x_train[num_cols])
x_test[num_cols]=scaler.transform(x_test[num_cols])
```

<h3> Linear Regression </h3>


```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test,y_test)*100
```

<h3> Decision Tree </h3>

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
dt.score(x_test,y_test)*100
```

<h3> KNN classifier </h3>

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
```

```python
knn.score(x_test,y_test)*100
```

```python
## Random Forest Classifier 

from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=200, random_state=42)
rf.fit(x_train,y_train)
```

```python
## Random forect accuracy

rf_acc= rf.score(x_test,y_test)
print("Random forest accuracy: ",rf_acc)
```

```python
## random forest with hypertuning 
rf = RandomForestClassifier(
    n_estimators=500,        
    max_depth=20,           
    min_samples_split=4,
    min_samples_leaf=2,     
    max_features="sqrt",  
    random_state=42
)

rf.fit(x_train, y_train)

y_pred = rf.predict(x_test)
from sklearn.metrics import accuracy_score
print("New Accuracy:", accuracy_score(y_test, y_pred))
```
