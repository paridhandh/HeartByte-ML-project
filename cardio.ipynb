```python
## check if missing values
df.isnull().sum()
```

```python
## drop id colums
df.drop(columns=['id'],inplace=True)
df
```

```python
## converts age to years from days
df['age_years']=(df['age']/365).astype(int)
df.drop(columns=['age'],inplace=True)
df
```

```python
## remove wrong medical data ---- outliers

## ap_lo> ap_hi --- not valid
df= df[df['ap_lo']<=df['ap_hi']]
df
```

```python
## height <100  or >220--- unrealistic 
df=df[(df['height']>100) & (df['height']<220)]
df
```

```python

## weight < 30 or >200
df=df[(df['weight']>30) & (df['weight']<200)]
df
```

```python
### calculate BMI
## BMI = weight / (height/100)^2
df['BMI']=df['weight'] / ((df['height']/100)**2)
df
```

```python
from  sklearn.model_selection import train_test_split
## seperate input and output
## x = testing data
## y = target data

x = df.drop(columns=['cardio'],axis=1)  ## all data except cardio column is training data
y = df['cardio']  ## cardio is ourtarget data to check the correctness of model

x_train, x_test , y_train, y_test = train_test_split(
    x,y,test_size=0.2 , random_state=42,
)
```

```python
x_train
```

```python
df
```

```python
num_cols=['height','weight','ap_hi','ap_lo','age_years','BMI']
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train[num_cols])
```

```python
x_train[num_cols]=scaler.transform(x_train[num_cols])
x_test[num_cols]=scaler.transform(x_test[num_cols])
```

```python
### LogisticRegression

from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(max_iter=1000, solver="liblinear")
log_reg.fit(x_train, y_train)
```

```python
## test accuracy of LogisticRegression
log_acc = log_reg.score(x_test, y_test)
print("Logistic Regression Accuracy:", log_acc)
```

```python
## linear regression
from sklearn import linear_model
lr = linear_model.LinearRegression()
lr.fit(x_train, y_train)
```

```python
## test accuracy of linear regression
lr_acc = lr.score(x_test,y_test)
print("Linear Regression Accuracy: ", lr_acc)
```

```python
## KNN classifier
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
```

```python
## check accuracy of KNN

knn_acc = knn.score(x_test,y_test)
print("KNN accuracy: ", knn_acc)
```

```python
## Decision Tree
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
## check accuracy of Decision Tree
dt_acc = dt.score(x_test,y_test)
print("Desicion Tree Accuracy: ",dt_acc)
```

```python
## Random Forest Classifier 

from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=200, random_state=42)
rf.fit(x_train,y_train)
```

```python
## Random forect accuracy

rf_acc= rf.score(x_test,y_test)
print("Random forest accuracy: ",rf_acc)
```

```python
## random forest  hypertuning 
rf = RandomForestClassifier(
    n_estimators=500,        
    max_depth=20,           
    min_samples_split=4,
    min_samples_leaf=2,     
    max_features="sqrt",  
    random_state=42
)

rf.fit(x_train, y_train)

y_pred = rf.predict(x_test)
from sklearn.metrics import accuracy_score
print("New Accuracy:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

# rf = RandomForestClassifier(
#     n_estimators=300,
#     max_depth=None,
#     random_state=42,
#     n_jobs=-1
# )
rf = RandomForestClassifier(
    n_estimators=600,
    max_depth=12,
    min_samples_split=20,
    min_samples_leaf=10,
    max_features='sqrt',
    class_weight='balanced',
    random_state=42,
    n_jobs=-1
)

rf.fit(x_train,y_train)
y_pred = rf.predict(x_test)

rf.score(x_test,y_test)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=3,
    random_state=42
)

gb.fit(x_train, y_train)
y_pred_gb = gb.predict(x_test)

print("Gradient Boosting Accuracy:", accuracy_score(y_test, y_pred_gb))
```
